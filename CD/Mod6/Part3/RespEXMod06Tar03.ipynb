```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook
df = pd.read_csv('/home/aleubuntu/CientistaDados/CientistaDados/CD/Mod6/Part1/sinasc_RO_2019.csv')
df.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[(df['APGAR5'] <= 3), 'resp_cat'] = 'asfixia grave'
df.loc[(df['APGAR5'] > 3) & (df['APGAR5'] <= 5),'resp_cat'] = 'asfixia moderada'
df.loc[(df['APGAR5'] > 5) & (df['APGAR5'] <= 7),'resp_cat'] = 'asfixia leve'
df.loc[(df['APGAR5'] > 7), 'resp_cat'] = 'normal'
```

```python
df
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python

tempogestacao = [['Menos de 22 semanas', '22 a 27 semanas', '28 a 31 semanas'],
                 ['32 a 36 semanas', '37 a 41 semanas', '42 semanas e mais']]

fig, axes = plt.subplots(2, 3)

for i in range(len(tempogestacao)):
    for j in range(len(tempogestacao[i])):
        sns.countplot(ax = axes[i, j]
                      , x = 'APGAR5'
                      , hue = 'SEXO'
                      , data = df[df['GESTACAO'] == tempogestacao[i][j]]
                     ).legend(loc = 'upper left').set(title=tempogestacao[i][j])

plt.subplots_adjust(wspace = 0.15, hspace = 0.15)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tab_cruzada = pd.crosstab(df['GESTACAO'], df['APGAR5'])

tab_cruzada
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tab_cruzada_per_linha = tab_cruzada.div(tab_cruzada.sum(axis = 1), axis = 0)
tab_cruzada_per_linha
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
tab_cruzada_per_linha.plot.bar()
```

```python
#Segundo analise preliminar, o gráfico acima corrobora com as classificações dos médicos, pois apresenta uma maior probabilidade de asfixia abaixo de 37 semanas.
```

#### 7) Veja se há relação entre APGAR5 e o peso do bebê. Isso pode ser feito de diversas formas.
```python
lista_cat = [['normal', 'asfixia leve'],
                 ['asfixia moderada', 'asfixia grave']]

plt.rc('figure', figsize = (20, 10))

fig, axes = plt.subplots(2, 2)

for i in range(len(lista_cat)):
    for j in range(len(lista_cat[i])):
        sns.histplot(ax = axes[i, j]
                     , x = 'PESO'
                     , bins = 10
                     , data = df[df['resp_cat'] == lista_cat[i][j]]
                    ).set(title=lista_cat[i][j])

plt.subplots_adjust(wspace = 0.25, hspace = 0.25)
print(f' É possível observar que bebês com peso abaixo de 3kgs tem maior propensão a problemas respiratórios')
```
